# Defining function to pull MU baseball data

* gives global "db_names" and "table_data"
* also gives database name map with "name_map" dictionary

```python
import requests 
from bs4 import BeautifulSoup 
from pprint import pprint
import pandas as pd
import numpy

def pull_MU(year):
    url = 'https://floridagators.com/sports/baseball/stats/'+str(year)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html")
    table = soup.table
    
    trs = table.find_all('tr')
    header_row = trs[0]
    
    names = []
    for column in header_row.find_all('th'):
        names.append(column.text)
    names.remove('Bio Link')
    
    global db_names
    global table_data
    table_data = []
    db_names = []
    for column in names:
        db_names.append(column.lower())
    table_data = table.find_all('tr')
    table_data = table_data[1:]
    
    global name_map
    names_full = ['Player Number', 'Name','Batting Average','On-base Plus Slugging',
             'Games Played','Games Started','At-bat','Run','Hit','Double','Triple','Home Run',
             'Runs Batted In','Total Bases','Slugging Percentage','Walk','Hit-by-pitch','Strikeout',
             'Ground Into Double Play','On-base Percentage','Sacrifice Fly','Sacrifice Bunt',
             'Stolen bases allowed','Stealing attempts', 'Year']
    db_names[4] = 'gp'
    db_names[21] = 'sb'
    db_names.insert(22,'att')
    db_names.insert(5, 'gs')
    db_names.append('year')
    name_map = dict(zip(db_names, names_full))
```

# Insert data into Pandas Dataframe

* Combines all df into one
* Saves as HDF5 file

```python
df2007 = pd.DataFrame(uf2007)
df2008 = pd.DataFrame(uf2008)
df2009 = pd.DataFrame(uf2009)
df2010 = pd.DataFrame(uf2010)
df2011 = pd.DataFrame(uf2011)
df2012 = pd.DataFrame(uf2012)
df2013 = pd.DataFrame(uf2013)
df2014 = pd.DataFrame(uf2014)
df2015 = pd.DataFrame(uf2015)
df2016 = pd.DataFrame(uf2016)
df2017 = pd.DataFrame(uf2017)
df2018 = pd.DataFrame(uf2018)
df2019 = pd.DataFrame(uf2019)
```

```python
mu_baseball = pd.concat([df2007, df2008, df2009, df2010, df2011, 
                         df2012, df2013, df2014, df2015, 
                         df2016, df2017, df2018, df2019])
```

```python
filepath = "./dataframes/baseball.h5"
mu_baseball.to_hdf(filepath, key = 'uf')
```
